```python
import pandas as pd
```

```python
data = pd.read_csv('Pokemon.csv')
data.head()
```

```python
data.info
```

```python
data.describe()
```

```python
data.sort_values("Defense",ascending=False)
```

```python
defense_mean = data["Defense"].mean()

def defense_val_set(value):
    if value > defense_mean:
        return "High defense"
    elif value == defense_mean:
        return "Neutral defense"
    else:
        return "Low defense"
    
hp_mean = data["HP"].mean()

def hp_val_set(value):
    if value > hp_mean:
        return "High HP"
    elif value == hp_mean:
        return "Neutral HP"
    else:
        return "Low HP"
data["Defense_high_low"] = data["Defense"].apply(defense_val_set)
data["HP_high_low"] = data["HP"].apply(hp_val_set)
data.head()
```

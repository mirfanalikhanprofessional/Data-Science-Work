```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('details_test.csv')
data.head()
```

```python
# One HOT Encoding
data['Production Country'].replace(np.nan, "United States", inplace=True)
```

```python
# Show the most number of tv-shows and movies released by each country in descending orderd
data['Production Country'].value_counts().sort_values(ascending=False)
```

```python
subdatap1 = data.groupby('Content Type')
subdatap2 = subdatap1['Production Country'].value_counts()
subdataframe = pd.DataFrame(subdatap2)
subdataframe
```

```python
# Also plot a countplot of each country.
import seaborn as sb
plt.figure(figsize=(300,300))
# %matplotlib inline
sb.countplot(x = 'Production Country', data = data, )
plt.show()
```
